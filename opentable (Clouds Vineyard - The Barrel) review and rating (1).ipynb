```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url_1=[]

for i in range(1,6):
    url="https://www.opentable.com/r/clouds-vineyard-the-barrel-bald-knob?originId=3d5a593d-d99e-465d-9e26-35ec1cada51d&corrid=3d5a593d-d99e-465d-9e26-35ec1cada51d&avt=eyJ2IjoyLCJtIjoxLCJwIjowLCJzIjowLCJuIjowfQ&page={}&sortBy=newestReview".format(i)
    url_1.append(url) 
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/601.3.9 (KHTML, like Gecko) Version/9.0.2 Safari/601.3.9'}

items=[]

for rev in url_1:
    search_response=requests.get(rev,headers=headers)
    search_response.text
    soup =BeautifulSoup(search_response.text,"lxml")
    for item in soup.find_all('li', attrs = {'class': 'mbK0eco4h3AaJgKLBhb9'}):
        items.append(item) 
```

```python
review_2=[]
rating_2=[]

for i in items: 
    rating=i.find('span',{'class': 'Q2DbumELlxH4s85dk8Mj'}).text.strip()
    review=i.find('span',{'class': 't9JcvSL3Bsj1lxMSi3pz h_kb2PFOoyZe1skyGiz9 vF9xoC79LKrlQXgOuzNk'}).text.strip()
    review_2.append(review)
    rating_2.append(rating)

rating_2
```

```python
# Taking 1 if rating is 3 or more than that and 0 if it is below 3

a = 0
while a < len(rating_2):
  
    # replace hardik with shardul
    if rating_2[a] >= '3':
        rating_2[a] = '1'
    else:
        rating_2[a] = '0'
  
    a += 1
  
# print list
print(rating_2)
```

```python
data={'Review':review_2,'Rating':rating_2}
```

```python
df=pd.DataFrame(data)
df
```
